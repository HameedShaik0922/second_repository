```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import traceback
import statsmodels.api as sm
import numpy as np
from scipy import stats
# import statsmodels
```

```python
df = pd.read_csv('elasticity.csv', encoding = "ISO-8859-1")
df.head()
```

```python
products = df.PRODUCT_NAME.unique()
```

```python
df_columns = ["name", "price_elasticity", "price_mean", "quantity_mean", "intercept", "t_score", "slope", "coefficient_pvalue", "r_square"]

name = []
price_elasticity = []
price_mean = []
quantity_mean = []
intercept = []
t_score =[]
slope = []
coefficient_pvalue = []
r_square = []

for product in products:
    print(product)
    df1= df[df['PRODUCT_NAME'] == product]
    x = df1['PRICE'].to_numpy()
    y = df1['AVG_QTY'].to_numpy()

    X = sm.add_constant(x)
    model = sm.OLS(y,X).fit()
    
    product_name = product
    p_elasticity = model.params[1] * np.mean(x) / np.mean(y)
    p_mean = np.mean(x)

    q_mean = np.mean(y)
    p_intercept =  model.params[0]
    tscore = model.tvalues[0]

    p_slope = model.params[1]
    p_coefficient = model.f_pvalue
    rsquare = model.rsquared
    
    name.append(product_name)
    price_elasticity.append(p_elasticity)
    price_mean.append(p_mean)

    quantity_mean.append(q_mean)
    intercept.append(p_intercept)
    t_score.append(tscore)

    slope.append(p_slope)
    coefficient_pvalue.append(p_coefficient)
    r_square.append(rsquare)

    # X = np.array(X, dtype = float)

```

```python
df_final = pd.DataFrame(list(zip(name, price_elasticity, price_mean, quantity_mean, intercept, t_score, slope, coefficient_pvalue, r_square)), columns = df_columns)
display(df_final)
```
